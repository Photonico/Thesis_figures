```python
# font test

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Sample text for shape comparison
sample_text = "Qgar"

# Candidate fonts to compare
# Keep the list long; unavailable fonts will be filtered out automatically.
candidate_fonts = [
    # Computer Modern / TeX-like
    "cmr10",
    "cmb10",
    "cmmi10",
    "cmss10",
    "cmtt10",

    # STIX
    "STIXGeneral",
    "STIX Two Text",
    "STIX Two Math",

    # DejaVu
    "DejaVu Serif",
    "DejaVu Serif Display",
    "DejaVu Sans",
    "DejaVu Sans Display",
    "DejaVu Sans Mono",

    # Common serif / book fonts
    "Charter",
    "Palatino",
    "Times",
    "Times New Roman",
    "Georgia",
    "Baskerville",
    "Big Caslon",
    "Bodoni 72",
    "Bodoni 72 Oldstyle",
    "Bodoni 72 Smallcaps",
    "Cochin",
    "Didot",
    "Hoefler Text",
    "Iowan Old Style",
    "Marion",
    "New York",
    "PT Serif",
    "PT Serif Caption",
    "Publico Text",
    "Athelas",
    "Canela Text",
    "Plantagenet Cherokee",

    # Slab / sturdy / typewriter-like
    "Rockwell",
    "Superclarendon",
    "American Typewriter",
    "Courier",
    "Courier New",
    "Andale Mono",

    # Humanist / readable text fonts
    "Optima",
    "Skia",
    "Lucida Grande",
    "Geneva",
    "Kefa",

    # Sans fonts that may still be useful for glyph comparison
    "Helvetica",
    "Helvetica Neue",
    "Arial",
    "Arial Narrow",
    "Avenir",
    "Avenir Next",
    "Avenir Next Condensed",
    "Gill Sans",
    "Futura",
    "PT Sans",
    "Microsoft Sans Serif",
    "Trebuchet MS",
    "Verdana",
    "Tahoma",
    "Seravek",
    "Proxima Nova",

    # Monos / coding fonts that may still be interesting
    "Menlo",
    "Monaco",
    "PT Mono",
    "Photonico Code",
    "FiraCode Nerd Font",
    "FiraCode Nerd Font Mono",
    "FiraCode Nerd Font Propo",
    "GoMono Nerd Font",
    "GoMono Nerd Font Mono",
    "GoMono Nerd Font Propo",
    "MesloLGS NF",

    # Decorative but still Latin-readable
    "Academy Engraved LET",
    "Apple Chancery",
    "Copperplate",
    "Herculanum",
    "Luminari",

    # CJK serif / Mincho / Ming / Myungjo-like fonts with Latin glyphs
    "Songti SC",
    "YuMincho",
    "Hiragino Mincho ProN",
    "BIZ UDMincho",
    "PCMyungjo",
    "AppleMyungjo",
    "Nanum Myeongjo",
    "LiSong Pro",
    "SimSong",
    "STFangsong",
    "Toppan Bunkyu Mincho",
    "Toppan Bunkyu Midashi Mincho",

    # Other CJK fonts that may still be interesting for Latin letterforms
    "Klee",
    "Kokonor",
    "Kai",
    "Kaiti SC",
    "BiauKaiHK",
]
# Get fonts actually available in the local Matplotlib font database
available_fonts = set(fm.get_font_names())
fonts_to_show = [name for name in candidate_fonts if name in available_fonts]

print("Fonts that will be compared:")
for name in fonts_to_show:
    print(name)

# Figure layout
n = len(fonts_to_show)
fig_height = max(6, 0.55 * n + 1.0)
fig, ax = plt.subplots(figsize=(12, fig_height), dpi=220)

ax.set_xlim(0, 1)
ax.set_ylim(0, n + 1)
ax.axis("off")

# Title
ax.text(0.02, n + 0.4,
        "Font comparison: Q g a r",
        fontsize=16,
        fontfamily="DejaVu Sans",
        va="center")
# Draw one row per font
for i, font_name in enumerate(fonts_to_show):
    y = n - i
    # Left: font name
    ax.text(0.02, y,
            font_name,
            fontsize=12,
            fontfamily="DejaVu Sans",
            va="center")
    # Right: sample text
    ax.text(0.36, y,
            sample_text,
            fontsize=30,
            fontfamily=font_name,
            va="center")
# Save and show
plt.tight_layout()
# plt.savefig("font_compare", bbox_inches="tight")
plt.show()
```
